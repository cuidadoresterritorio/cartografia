```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./angry_birds.csv')
print(df.shape)
df.head()
```

```python
sns.countplot(data=df, y='Localidad')
```

```python
filtro_name = df[df['Localidad'] == 'hato grande']
filtro_name.shape
```

```python
sns.countplot(data=filtro_name, y='Especies')
plt.figure(figsize=(12, 6))
plt.title('HATO GRANDE')

```

```python
sns.countplot(data=filtro_name, y='Especies', hue='Conteo');
plt.figure(figsize=(12, 24))
```

```python
plt.figure(figsize=(12, 24))
sns.countplot(data=df, y='Especies', hue='Localidad', dodge=True)
```

```python
sns.displot(data=df, x='Localidad', hue='GENERO', multiple='dodge')
```
